```python
import pandas as pd
import os, glob as g
files = g.glob(os.path.join(os.getcwd(), "*.csv"))
print(f"Files: {len(files)}, {files}")
```

```python
holder = dict()
for file in files:
    name = file.split("\\")[-1].strip()
    holder[name] = pd.read_csv(file)
    
    
```

```python
collect = pd.DataFrame()
for k, v in holder.items():
    print(k)
    print(len(v.columns))
    print(v.columns.tolist())
    print(v.shape)
    if len(v.columns) == 6:
        collect = pd.concat([collect, v])
```
